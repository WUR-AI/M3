```python
import datamart_profiler
import pandas as pd
from pathlib import Path

# Walk up from cwd until we find the repo data folder (works from any notebook subdir)
BASE = Path.cwd().resolve()
for candidate in [BASE, *BASE.parents]:
    csv_path = candidate / "data" / "biota_dataset" / "samples.csv"
    if csv_path.exists():
        BASE = candidate
        break
else:
    raise FileNotFoundError(
        "Could not find data/biota_dataset/biota.csv. "
        "Expected it under the metadata_agent repo root."
    )

data_frame = pd.read_csv(csv_path)
metadata = datamart_profiler.process_dataset(data_frame)
```
